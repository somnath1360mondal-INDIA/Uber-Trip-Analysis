```python
import pandas as pd
```

```python
df_uber = pd.read_parquet('Drafts/df_uber_parquet.parquet')
```

```python
df_uber.columns
```

```python
Dim_Rate = df_uber[['RatecodeID']].drop_duplicates().copy()
```

```python
Dim_Rate
```

```python
Dim_Rate = Dim_Rate.rename(columns={'RatecodeID':'rate_code_id'})
```

```python
Dim_Rate['rate_code_name'] = Dim_Rate['rate_code_id'].map({
    1: 'Standard Rate',
    2: 'JFK Airport Flat Fare',
    3: 'Newark Airport',
    4: 'Nassau / Westchester',
    5: 'Negotiated Fare',
    6: 'Group Ride'
})
```

```python
Dim_Rate.to_parquet('Final/Dim_Rate.parquet',index=False)
```
